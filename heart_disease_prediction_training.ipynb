```python
import keras
```

```python
from keras.models import Sequential
model = Sequential()
```

```python
from keras.layers import Dense
```

```python
import pandas as pd
dt = pd.read_csv('/content/heart.csv')
dt.head(5)
```

```python
dt.corr()
```

```python
dt.drop(['chol','fbs'],axis=1)
```

```python
dt.head()
```

```python
x = dt.iloc[:,0:11]
y = dt['target']
```

```python
x
```

```python
y
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
print(x.shape,y.shape,x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
model.add(Dense(units = 11,activation='relu'))
model.add(Dense(units=5, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.build(input_shape=(None,11))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=8,steps_per_epoch=x_train.shape[0]//8)
```

```python
model.save('heart.h5')
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
import numpy as np

# Convert predicted probabilities to binary class labels
# We use a threshold of 0.5. If the predicted probability is >= 0.5, it's class 1, otherwise class 0.
y_pred_binary = (y_pred >= 0.5).astype(int)

# Calculate accuracy using the binary predictions
accuracy_score(y_test, y_pred_binary)
```
